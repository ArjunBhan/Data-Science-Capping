```python
#Imports and Downloads
import pandas as pd
import nltk
import re
from nltk.stem import WordNetLemmatizer 
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('vader_lexicon')
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer


# setup to list and objects needed for the code
Tweet_List_Sentiment = []
Clean_Tweet_List = []
Positive_Negative=[]
value=" "
count=0
ps = PorterStemmer()
lemmatizer=WordNetLemmatizer()
sia=SentimentIntensityAnalyzer()
# Import data

#Use the tweets (1).csv from the Neural Network and Text Analysis Folder
Tweet_DF=pd.read_csv(r'/tweets (1).csv',names=["Number","Handle","ID","Date","Tweet"])
Tweet_List=Tweet_DF['Tweet'].tolist()
Big_Handle_List=Tweet_DF['Handle'].tolist()
Date_Time=Tweet_DF['Date'].tolist()
ID_Lis=Tweet_DF['ID'].tolist()
Play_ID_Lis=Tweet_DF['Number'].tolist()
# process tweets
for t, tweet in enumerate(Tweet_List):
  # tokenize and lower
  clean_tweet_bag = [t.lower() for t in nltk.word_tokenize(tweet)]
  # remove non alphanumeric stuff like punctuation and numbers
  clean_tweet_bag = [t for t in clean_tweet_bag if t.isalnum()]
  # process tweet handles
  # remove stop words
  clean_tweet_bag = [t for t in clean_tweet_bag if t not in nltk.corpus.stopwords.words("english")]
  # stem tokens
  clean_tweet_bag = [ps.stem(t) for t in clean_tweet_bag]
  # lemmatize tokens
  clean_tweet_bag = [lemmatizer.lemmatize(t) for t in clean_tweet_bag]
  # add cleaned tweet text to list
  Clean_Tweet_List.append(clean_tweet_bag)
  # calculate tweet sentiment
  tweet_sentiment = sia.polarity_scores(" ".join(Clean_Tweet_List[t]))
  neg=tweet_sentiment['neg']
  pos=tweet_sentiment['pos']
  Positive_Negative.append([neg, pos])
```

```python
#Counter for postive and negative tweets
pos_count = 0
neg_count = 0
Pre_Handle=Big_Handle_List[0]
for Hand_Num,handle in enumerate(Big_Handle_List):
  if handle !=Pre_Handle and Hand_Num != 0:
    pos_count = 0
    neg_count = 0
  pos_count +=(Positive_Negative[Hand_Num][1])
  neg_count +=(Positive_Negative[Hand_Num][0]) 
  Pre_Handle=handle
```

```python
#Makes sure time stamps not included
Date_List=[]
for num,date in enumerate(Date_Time):
  Post_Date,Post_Time=Date_Time[num].split(' ')
  Date_List.append(Post_Date)

```

```python
#Sets up a list containing all relvant information of the tweets.
pos_count = 0
neg_count = 0
Cur_Date=Date_List[0]
pos_neg_count=0
ID_Date_Lis=[]
for Hand_Num,handle in enumerate(Big_Handle_List):
  if Cur_Date !=Date_List[Hand_Num] and Hand_Num != 0:
    if pos_neg_count !=0:
      ID_Date_Lis.append([handle,Play_ID_Lis[Hand_Num],pos_neg_count,Date_List[Hand_Num]])
    pos_count = 0
    neg_count = 0
  pos_count +=(Positive_Negative[Hand_Num][1])
  neg_count +=(Positive_Negative[Hand_Num][0]) 
  pos_neg_count=pos_count-neg_count
  Cur_Date=Date_List[Hand_Num]
```
